```python
import tensorflow as tf
import numpy as np

R = tf.constant(6371.0, dtype=tf.float64)  # Radius of the Earth in kilometers

# Mercator projection
def mercator_projection(lat, lon):
    x = R * lon
    y = R * tf.math.log(tf.math.tan(np.pi / 4 + lat / 2))
    return x, y
```

The distortion in the Mercator projection is calculated in terms of scale factors.
The scale factor in the latitude direction  

k(ϕ)=sec(ϕ)

```python
def calculate_distortion(lat):
    k = 1 / tf.math.cos(lat)
    return k
```

```python
latitudes = tf.constant(np.linspace(-np.pi / 2, np.pi / 2, num=1000), dtype=tf.float64)
longitudes = tf.constant(np.zeros_like(latitudes), dtype=tf.float64)  # zero longitude

x, y = mercator_projection(latitudes, longitudes)

distortion = calculate_distortion(latitudes)

# Calculate results directly using eager execution
x_val = x.numpy()
y_val = y.numpy()
distortion_val = distortion.numpy()

# Output the results
print("X coordinates (Mercator projection):", x_val)
print("Y coordinates (Mercator projection):", y_val)
print("Distortion factors:", distortion_val)
```
