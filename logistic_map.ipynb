```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x):
    """Computes the next value in the logistic map sequence."""
    return r * x * (1 - x)

def plot_logistic_map(r_values, initial_x=0.5, iterations=500, transient=100):
    """Plots the bifurcation diagram of the logistic map."""

    x_values = []
    r_results = []

    for r in r_values:
        x = initial_x
        results = []
        for _ in range(iterations):
            x = logistic_map(r, x)
            results.append(x)

        x_values.extend(results[transient:])
        r_results.extend([r] * (iterations - transient))

    plt.figure(figsize=(10, 6))
    plt.plot(r_results, x_values, ls='', marker=',', markersize=0.1)
    plt.xlabel("r")
    plt.ylabel("x")
    plt.title("Bifurcation Diagram of the Logistic Map")
    plt.xlim(min(r_values), max(r_values))
    plt.ylim(0, 1)
    plt.show()

if __name__ == "__main__":
    r_values = np.linspace(2.5, 4.0, 1000)
    plot_logistic_map(r_values)


    r_specific = 3.8
    x_series = [0.5]
    for _ in range(200):
        x_series.append(logistic_map(r_specific, x_series[-1]))
    plt.figure()
    plt.plot(x_series)
    plt.title(f"Time series for r = {r_specific}")
    plt.xlabel("Iteration")
    plt.ylabel("x")
    plt.show()

    def plot_cobweb(r, x0, n):
        x = np.linspace(0, 1, 1000)
        y = logistic_map(r, x)

        plt.figure()
        plt.plot(x, y, label="f(x)")
        plt.plot(x, x, label="y=x")

        xn = x0
        for i in range(n):
            yn = logistic_map(r, xn)
            plt.plot([xn, xn], [xn, yn], 'r-', alpha=0.5)
            plt.plot([xn, yn], [yn, yn], 'r-', alpha=0.5)
            xn = yn
        plt.title(f"Cobweb diagram for r = {r}, x0 = {x0}")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.legend()
        plt.show()

    plot_cobweb(3.8, 0.2, 10)
```
