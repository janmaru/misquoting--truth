```python
import math

# using PNT
def approximate_prime_probability(n):
    if n < 2:
        return 0.0
    return 1 / math.log(n)

numbers_to_evaluate = [11, 13, 17, 15, 16, 18]

probabilities = {num: approximate_prime_probability(num) for num in numbers_to_evaluate}

total_probability = sum(probabilities.values())
normalized_probabilities = {num: prob / total_probability for num, prob in probabilities.items()}

print("Normalized probabilities:")
for num, prob in normalized_probabilities.items():
    print(f"Number: {num}, Probability: {prob:.4f}")
```

```python
import math

# Function to approximate prime probability using PNT
def approximate_prime_probability(n):
    if n < 2:
        return 0.0  # No primes below 2
    return 1 / math.log(n)

# Define numbers to evaluate
numbers_to_evaluate = [11, 13, 17, 15, 16, 18]

# Compute raw probabilities based on PNT
probabilities = {num: approximate_prime_probability(num) for num in numbers_to_evaluate}

# Normalize probabilities
total_probability = sum(probabilities.values())
normalized_probabilities = {num: prob / total_probability for num, prob in probabilities.items()}

# Display results
print("Normalized probabilities:")
for num, prob in normalized_probabilities.items():
    print(f"Number: {num}, Probability: {prob:.4f}")
```

```python
import matplotlib.pyplot as plt

numbers = list(normalized_probabilities.keys())
probabilities = list(normalized_probabilities.values())

plt.bar(numbers, probabilities)

plt.xlabel("Numbers")
plt.ylabel("Normalized Probabilities")
plt.title("Normalized Probabilities of Numbers")
plt.xticks(numbers)

plt.show()
```
